### Day 3: Advanced Slicing and dicing

#### Load packages

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### Read csv and check type of data in the dataframe

```python
df = pd.read_csv('FL_insurance_sample.csv')

df.head()
```

### Select only those rows where line is 'Residential'

```python
residential_line = df[df['line']=='Residential']
```

#### Lets check the first 5 rows

```python
residential_line[:5]
```

#### To select only those rows where line is 'Residential' and construction is 'Wood'

#### Shwoing first 5 rows from the result set directly

```python
residential_line[residential_line['construction']=='Wood'][:5]
```

#### Select above data in a single step

```python
df[(df['line'] == 'Residential') & (df['construction'] == 'Wood')][:5]
```

#### We can also select few columns at single go

```python
df[(df['line'] == 'Residential') & (df['construction'] == 'Wood')][['statecode', 'line', 'construction']][:5]
```

### So to check which county has most residential claims

```python
residential_line['county'].value_counts()
```

#### Check the type of construction in residential

```python
residential_line['construction'].value_counts()
```

#### Check what percentage of total claims are residential claims for each county

```python
total_claims = df['county'].value_counts()

residential_claims = residential_line['county'].value_counts()

(residential_claims/total_claims).astype(float)
```
