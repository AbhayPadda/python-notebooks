## Day 2: Selecting Columns and slicing Datasets

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (15, 5)
```

### Read CSV

```python
df = pd.read_csv("FL_insurance_sample.csv")
```

#### Show first 5 rows in the dataframe

```python
df.head(5)
```

#### We can similarly show last 5 rows in the dataframe

```python
df.tail(5)
```

#### Alternatively, we can use the below method as well to select first 5 rows

```python
df[:5]

# To select last 5 rows, we can use:
## df[-5:] 
```

#### We can similary combine this to get first 5 rows for a column

```python
df['county'][:5]

## df[:5]['county'] : This will also give the same output 
```

#### To select multiple columns using name

```python
df[['statecode', 'county']]
```

#### To check the most frequently occurring county (the result is by default sorted in descending order)

```python
df['county'].value_counts()
```

#### We can select top 5 counties using

```python
df['county'].value_counts()[:5]
```

#### And then plot a histogram

```python
df['county'].value_counts()[:5].plot(kind="bar")
```
