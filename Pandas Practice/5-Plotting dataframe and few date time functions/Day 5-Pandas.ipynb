### Day 5: Working with aggregate functions and date time manipulations

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### Read csv and plot Temp by month

```python
weather_2012_final = pd.read_csv('weather_2012.csv', index_col='Date/Time', parse_dates=True)
weather_2012_final['Temp (C)'].plot(figsize=(15, 6))
```

#### Check dataframe

```python
weather_2012_final.head()
```

Plot Temp with Date Time

##### Look's like temperature is maximum between mid June to mid Aug

```python
weather_2012_final["Temp (C)"].plot(figsize=(15, 5))
```

#### Drop columns where any value is 1

```python
weather_2012_final = weather_2012_final.dropna(axis=1, how='any')
weather_2012_final[:5]
```

#### Extract hour from the data/time column and save in a new column

```python
weather_2012_final.loc[:,'hour'] = weather_2012_final.index.hour
```

#### Group by hour of the day and plot for the mean value by hour.

```python
weather_2012_final.groupby('hour').mean().plot(figsize=(15,5))
```
